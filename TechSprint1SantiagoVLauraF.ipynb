```python
import pandas as pd

db = pd.read_csv('db-sprint1-sti.csv')
```

```python
print(db.head())
```

```python
def orderedDB(dataframe, column, organization):
    if organization == 'ascendente':
        activeOrg = True
    elif organization == 'descendente':
        activeOrg = False
    else:
        return ValueError('Error en el campo de "ascendete" o "descendente"')

    dbOrdenada = dataframe.sort_values(by=column, ascending=activeOrg).reset_index(drop=True)

    return dbOrdenada

# Solicitar al usuario la columna y dirección del orden
column_selection = input('Escribe el nombre de la columna que quieres ordenar: ')
organize_selection = input('Quieres ordenarla de manera "ascendente" o "descendente": ')

# Utilizar la función con los datos proporcionados por el usuario
newDB = orderedDB(db, column_selection, organize_selection)
print(newDB)
```
